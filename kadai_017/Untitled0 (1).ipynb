```python
class Human:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def check_adult(self):
    if self.age >= 20:
      print("大人です。")

    else:
      print("大人ではありません。")

human_list = [
    Human("田中", 25),
    Human("佐藤", 18),
    Human("鈴木", 20),
    Human("高橋", 12)
]

# 2. リストの要素数分だけループしてメソッドを呼び出す
for person in human_list:
    person.check_adult()

```
