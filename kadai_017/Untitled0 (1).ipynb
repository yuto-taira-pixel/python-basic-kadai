```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name} は大人です。")
        else:
            print(f"{self.name} は大人ではありません。")

humans = [
    Human("侍太郎", 25),
    Human("侍花子", 18),
    Human("侍一郎", 20)
]

for person in humans:
    person.check_adult()
```
